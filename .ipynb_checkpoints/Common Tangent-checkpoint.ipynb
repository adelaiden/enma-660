## Common Tangent Algorithm
The logic used to find the common tangent of two curves is shown in the method below. The program calculates instantaneous slopes and y-intercepts for each set of points initially provided, and then loops over both arrays of slopes and y-intercepts to find the minimum value for each curve. The program can only be used to find a single common tangent for now, but can be modified in the future to find multiple points of tangency.

```python
def find_common_tangent(x1, y1, x2, y2):    
    marr1 = np.array([]) #initialize slope and y intercept arrays
    marr2 = np.array([])
    barr1 = np.array([])
    barr2 = np.array([])
    
    for i in range(len(x1)-1): #calculate slope and y intercept for curve 1
        mval = (y1[i] - y1[i+1])/(x1[i]-x1[i+1])
        bval = y1[i] - mval*x1[i]
        marr1 = np.append(marr1, mval)
        barr1 = np.append(barr1, bval)
    
    for i in range(len(x2)-1): #calculate slope and y intercept for curve 2
        mval = (y2[i] - y2[i+1])/(x2[i]-x2[i+1])
        bval = y2[i] - mval*x2[i]
        marr2 = np.append(marr2, mval)
        barr2 = np.append(barr2, bval)
    
    dm = abs(marr1[0] - marr2[0])
    db = abs(barr1[0] - barr2[0])
    
    var = dm**2 + db**2
    for i in range(len(marr1)): #simultaneously minimize difference
        for j in range(len(marr2)):
            if var > (abs(marr1[i]-marr2[j]))**2 + (abs(barr1[i]-barr2[j]))**2:
                dm = abs(marr1[i]-marr2[j])
                db = abs(barr1[i]-barr2[j])
                var = dm**2 + db**2
                
                m1 = marr1[i]
                m2 = marr2[j]
                b1 = barr1[i]
                b2 = barr2[j]
                
                u1 = x1[i]
                u2 = x2[j]
                v1 = y1[i]
                v2 = y2[j]
                
    mfinal = (v1-v2)/(u1-u2)
    bfinal = v2 - mfinal*u2
    #print "Coordinates of points of tangency are: (%1.4f, %1.4f) "\
    #"and (%1.4f, %1.4f)" % (u1, v1, u2, v2)
    #print "A0 = ", bfinal, ", A1 = ", mfinal
    return mfinal, bfinal
```

# Homework 8
The following functions are used to find the points of tangency of the common tangent by finding the crossing points of the slope vs. y-intercept curves of the free energy graphs. The logic for the functions onSegment, orientation, and doIntersect is taken from the geeksforgeeks.org website. It checks to see whether two line segments intersect (and accounts for various cases).

```python
def onSegment(px, py, qx, qy, rx, ry):
    if (qx <= max(px, rx) and qx >= min(px, rx) and qy <= max(py, ry) and qy >= min(py, ry)):
        return True
    return False
 
#To find orientation of ordered triplet (p, q, r).
#The function returns following values
#0 --> p, q and r are colinear
#1 --> Clockwise
#2 --> Counterclockwise
def orientation(px, py, qx, qy, rx, ry):
    #See http://www.geeksforgeeks.org/orientation-3-ordered-points/ for details of below formula.
    val = (qy - py) * (rx - qx) - (qx - px) * (ry - qy)
 
    if val == 0: #colinear
        return 0
    if val > 0: #clockwise
        return 1
    if val < 0: #counterclockwise
        return 2

# The main function that returns true if line segment 'p1q1' and 'p2q2' intersect.
def doIntersect(p1x, p1y, q1x, q1y, p2x, p2y, q2x, q2y):
    #Find the four orientations needed for general and special cases
    o1 = orientation(p1x, p1y, q1x, q1y, p2x, p2y)
    o2 = orientation(p1x, p1y, q1x, q1y, q2x, q2y)
    o3 = orientation(p2x, p2y, q2x, q2y, p1x, p1y)
    o4 = orientation(p2x, p2y, q2x, q2y, q1x, q1y)
 
    #General case
    if (o1 != o2 and o3 != o4):
        return True
    #Special Cases
    #p1, q1 and p2 are colinear and p2 lies on segment p1q1
    if (o1 == 0 and onSegment(p1x, p1y, p2x, p2y, q1x, q1y)): 
        return True
    #p1, q1 and p2 are colinear and q2 lies on segment p1q1
    if (o2 == 0 and onSegment(p1x, p1y, q2x, q2y, q1x, q1y)):
        return True
    #p2, q2 and p1 are colinear and p1 lies on segment p2q2
    if (o3 == 0 and onSegment(p2x, p2y, p1x, p1y, q2x, q2y)): 
        return True
    #p2, q2 and q1 are colinear and q1 lies on segment p2q2
    if (o4 == 0 and onSegment(p2x, p2y, q1x, q1y, q2x, q2y)):
        return True

    return False #Doesn't fall in any of the above cases
```

```python
def find_intersections(x1, y1, x2, y2, plot_slopes=False):
    marr1 = np.array([]) #initialize slope and y intercept arrays
    marr2 = np.array([])
    barr1 = np.array([])
    barr2 = np.array([])
    
    for i in range(len(x1)-1): #calculate slope and y intercept for curve 1
        mval = (y1[i] - y1[i+1])/(x1[i]-x1[i+1])
        bval = y1[i] - mval*x1[i]
        marr1 = np.append(marr1, mval)
        barr1 = np.append(barr1, bval)
    
    for i in range(len(x2)-1): #calculate slope and y intercept for curve 2
        mval = (y2[i] - y2[i+1])/(x2[i]-x2[i+1])
        bval = y2[i] - mval*x2[i]
        marr2 = np.append(marr2, mval)
        barr2 = np.append(barr2, bval)

    idi = []
    idj = []
    selfi = []
    u1 = 0
    u2 = 0
    for i in range(len(barr1)-1):
        for j in range(len(barr2)-1):
            if doIntersect(marr1[i], barr1[i], marr1[i+1], barr1[i+1], marr2[j], barr2[j], marr2[j+1], barr2[j+1]):
                idi.append(i)
                idj.append(j)
    
    for i in range(len(marr1)-1):
        for j in range(len(marr1)-1):
            if i != j and i != j+1 and i != j-1 and selfi == []:
                if doIntersect(marr1[i], barr1[i], marr1[i+1], barr1[i+1], marr1[j], barr1[j], marr1[j+1], barr1[j+1]):
                    selfi.append(i)
                    u1 = x1[selfi]
                    u2 = x1[len(x1)-1 - selfi[0]] #len(x1)-1 - int((np.sqrt(selfi[0])))
            
    if plot_slopes:
        fig, ax = plt.subplots() #uncomment to visualize plot
        p1 = plt.plot(marr1, barr1, label='solid')
        p2 = plt.plot(marr2, barr2, label='liquid')
        p3 = plt.plot(marr1[idi], barr1[idi], 'ro')
        p4 = plt.plot(marr1[selfi], barr1[selfi], 'bo')
        
        plt.title(r'Slopes vs. Y-intercepts of Free Energy')
        plt.xlabel(r'$\mu$ x$_B$')
        plt.ylabel(r'dG/dx$_B$')
        legend = ax.legend(loc="best")
    
    print x1[idi], x2[idj], [marr1[selfi], barr1[selfi]] 
    return x1[idi], x2[idj], [marr1[selfi], barr1[selfi]], [u1, u2]
```

The program below plots the free energy curves for given parameters of T a0(l), and a0(s). It also plots a graph of the free energy curve slopes vs y-intercepts, with all points of intersections highlighted. Self-tangents (self-intersections) appear in blue, and common tangents (liquid-solid intersections) appear in red.

```python
a0s = 5000 #set values for constants
a0l = 0
t = 230
xb = np.linspace(.00000001, .99999999, num=100)
gunmixs = xb*4*(t-300) + (1-xb)*8*(t-400)
fs = gunmixs + a0s*(1-xb)*xb*xb + sc.R*t*((1-xb)*np.log(1-xb) + xb*np.log(xb))
fl = a0l*(1-xb)*xb + sc.R*t*((1-xb)*np.log(1-xb) + xb*np.log(xb))

me, be, se, si = find_intersections(xb, fs, xb, fl, plot_slopes=True)

fig, ax = plt.subplots()
plt.plot(xb, fs, label= 'solid')
plt.plot(xb, fl, label = 'liquid')

if se[0].size != 0:
    x_line = np.array(ax.get_xlim())
    y_line = se[1]+ se[0]*x_line
    labelstring = 'y = %1.2fx + %1.2f' % (se[0], se[1])
    plt.plot(x_line, y_line, '--', label='solid-solid self-tangent')

plt.title(r'Regular Solution Free Energy vs. x$_B$ at T = %1.0f K' %t)
plt.xlabel(r'x$_B$')
plt.ylabel(r'Free Energy (J)')
legend = ax.legend(loc="best")
```

This code will generate a phase diagram by plotting all points of tangency and self-tangency. Points of tangency are not plotted if the self-tangent is a lower free energy than the liquid (when the temperature becomes low enough). I have found that 200 points for both temperature and composition arrays will generate good-looking diagrams, but it takes about 3-4 minutes to generate the diagram. If you want to get a rough preview, use ~50 points for temperature and ~50 points for composition.

```python
a0s = 6000 #set values for constants
a0l = 1000

trange = np.linspace(10, 400, num=200) #change 'num' to increase or decrease number of points
xb = np.linspace(.00000001, .99999999, num=200)

x_1 = np.array([])
x_2 = np.array([])
x_3 = np.array([])
tarray = np.array([])
x_tarray = np.array([])
for t in trange:
    clear_output(wait=True)
    self_tangent = False
    gunmixs = xb*4*(t-300) + (1-xb)*8*(t-400) #change to change mu0
    fs = gunmixs + a0s*(1-xb)*xb + sc.R*t*((1-xb)*np.log(1-xb) + xb*np.log(xb)) #symmetric phase diagram
    fl = a0l*(1-xb)*xb + sc.R*t*((1-xb)*np.log(1-xb) + xb*np.log(xb))
    xsin, xlin, se, si = find_intersections(xb, fs, xb, fl)
    if se[0].size != 0:
        selft = se[1]+ se[0]*xb
        self_tangent = True
        for i in range(len(fl)):
            if (fl[i] - selft[i]) < 0:
                self_tangent = False
        if self_tangent:
            x_3 = np.append(x_3, [si[0], si[1]])
            x_tarray = np.append(x_tarray, [t,t])
        else:
            for i in range(len(xsin)):
                x_1 = np.append(x_1, xsin[i])
                x_2 = np.append(x_2, xlin[i])
                tarray = np.append(tarray, t)
    elif t > 300:
        for i in range(len(xsin)):
            x_1 = np.append(x_1, xsin[i])
            x_2 = np.append(x_2, xlin[i])
            tarray = np.append(tarray, t)
    print t

fig, ax = plt.subplots()
plt.scatter(x_1, tarray, c='black', s=10)
plt.scatter(x_2, tarray, c='black', s=10)
plt.scatter(x_3, x_tarray, c='blue', s=10)
plt.title(r'T vs. x$_B$ for a0(s) = %1.0f, a0(l) = %1.0f' %(a0s, a0l))
plt.xlabel(r'Composition x$_B$')
plt.ylabel(r'Temperature (K)')
```
