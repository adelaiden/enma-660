```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as sc
import math
```

```python
def high_temp_correction(gas, temp):
    '''Performs high-temp correction for the supplied gas name, given the new temperature.'''
    entropies = dict(O2=(205.152/2), SO2=(248.223/3), SO3=(256.77/4), P2O5=(403.96/14)) #gas phase entropy values
    pressures = dict(O2=0.2, SO2=10**-6, SO3=10**-6, P2O5=10**-8) #in atm
    s_e = entropies[gas]/sc.N_A
    h_corr = 3.5*sc.k*(temp - 298)
    p_new = pressures[gas]
    p_std = 1
    new_miu = (h_corr - temp*s_e + sc.k*temp*math.log(p_new/p_std))/sc.e
#     print 'new_miu: ', new_miu
#     print 'temp*s_e: ', temp*s_e
#     print 'h_corr: ', h_corr
#     print 'entropy term: ', temp*s_e/sc.e
    return new_miu
```

```python
temps = [298.15, 300, 400, 500, 600, 700, 800]
muO = []
for t in temps:
    muO.append(high_temp_correction('O2', t))
```

```python
muSO2 = []
muSO3 = []
muP2O5 = []
for t in temps:
    muSO2.append(high_temp_correction('SO2',t))
    muSO3.append(high_temp_correction('SO3',t))
    muP2O5.append(high_temp_correction('P2O5', t))
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.font_manager

f_size = 14
mpl.rcParams['xtick.labelsize']=f_size
mpl.rcParams['ytick.labelsize']=f_size
mpl.rcParams['axes.labelweight']='light'
mpl.rcParams['axes.labelsize']=f_size
mpl.rcParams['font.size']=f_size
mpl.rcParams['font.family']='sans-serif'
mpl.rcParams['font.sans-serif']='Arial'
mpl.rcParams['font.style']='normal'
mpl.rcParams['font.weight']='light'
mpl.rcParams['mathtext.fontset'] = 'custom'
mpl.rcParams['mathtext.rm'] = 'Arial'
mpl.rcParams["figure.figsize"] = (8,5)
```

```python
fig, ax = plt.subplots()
p1 = plt.plot(temps, muO, '-o', label=r'$\mu_O$')
p2 = plt.plot(temps, muSO3, '-o', label=r'$\mu_{SO_3}$')
#p3 = plt.plot(temps, muP2O5, label=r'$\mu_{P_2O_5}$')
p4 = plt.plot(temps, muSO2,'-o', label=r'$\mu_{SO_2}$')

ax.legend(loc='lower left')
plt.show()
```

Internal energy should have a linear relationship with chemical potential, so the overall trend of this correction equation looks correct.

```python
muO72 = []
for t in temps:
    muO72.append(high_temp_correction('O2', t))
```

```python
fig, ax = plt.subplots()
p1 = plt.plot(temps, muO, label=r'$\mu_O$, 5/2')
p2 = plt.plot(temps, muO72, label=r'$\mu_O$, 7/2')
ax.legend(loc='lower left')
plt.show()
```

This plot is the correction for O2, with 5/2, vs the correction for O2, with 7/2. Unfortunately it actually makes a pretty big difference at high temperatures, so we should get this one right.

$SO_2$ case study.

```python
exptT = [298.15, 300, 400, 500, 600, 700, 800]
exptG = [-300.125, -300.145, -300.971, -300.871, -300.305, -299.444, -298.37]
s_e = (248.223/3)/sc.N_A 

for i in range(len(exptG)):
    exptG[i] = exptG[i]/3

theorUstd = []
theorUtest = []
muSO2 = []
for t in exptT:
    ind = exptT.index(t)
    corr = sc.k*(t - 298.15)
    new_miu1 = (3.5*corr - t*s_e + sc.k*t*math.log(10**-6))/sc.e
    new_miu2 = (7*corr - t*s_e + sc.k*t*math.log(10**-6))/sc.e
    expt_miu = (exptG[ind]/sc.N_A + sc.k*t*math.log(10**-6))/sc.e
    theorUstd.append(new_miu1)
    theorUtest.append(new_miu2)
    muSO2.append(expt_miu)
```

```python
fig, ax = plt.subplots()
p1 = plt.plot(exptT, theorUstd, '-o', label=r'$\mu_{SO_2}$, 7/2')
p2 = plt.plot(exptT, theorUtest, '-o',label=r'$\mu_{SO_2}$, 7')
p3 = plt.plot(exptT, muSO2, '-o', label=r'$\mu_{SO_2}$, expt')
ax.legend(loc='lower left')
plt.show()
```

$SO_3$ case study.

```python
exptT = [298.15, 300, 400, 500, 600, 700, 800]
exptG = [-371.016, -370.862, -362.242, -352.668, -342.647, -332.365, -321.912]
s_e = (256.77/4)/sc.N_A 

for i in range(len(exptG)):
    exptG[i] = exptG[i]/4

theorUstd = []
theorUtest = []
muSO3 = []
for t in exptT:
    ind = exptT.index(t)
    corr = sc.k*(t - 298.15)
    new_miu1 = (3.5*corr - t*s_e + sc.k*t*math.log(10**-6))/sc.e
    new_miu2 = (10*corr - t*s_e + sc.k*t*math.log(10**-6))/sc.e
    expt_miu = (exptG[ind]/sc.N_A + sc.k*t*math.log(10**-6))/sc.e
    theorUstd.append(new_miu1)
    theorUtest.append(new_miu2)
    muSO3.append(expt_miu)
    
fig, ax = plt.subplots()
p1 = plt.plot(exptT, theorUstd, '-o', label=r'$\mu_{SO_3}$, 7/2')
p2 = plt.plot(exptT, theorUtest, '-o',label=r'$\mu_{SO_3}$, 10')
p3 = plt.plot(exptT, muSO3, '-o', label=r'$\mu_{SO_3}$, expt')
ax.legend(loc='lower left')
plt.show()
```

Results of these case studies: factoring vibrational and atomic degrees of freedom makes a difference. But neither are as accurate as experiment. Therefore when experimental values are available, use them.
