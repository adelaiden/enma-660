# Single-System

## O2 Evolution for Cathodes (LCO, LMO, LFP)

```python
from pymatgen import MPRester, Element, Composition
from pymatgen.phasediagram.maker import PhaseDiagram
from pymatgen.phasediagram.entries import PDEntry
from pymatgen.phasediagram.analyzer import PDAnalyzer
from pymatgen.analysis.reaction_calculator import ComputedReaction
import matplotlib.pyplot as plt
import sys
```

```python
def get_stability_window(chsys, formula, elem):
    '''chsys is the a list of strings of the elements in the chemical system, 
    formula is a string of the chemical compound, and elem is the reference element.'''
    cat = 0
    ano = 0
    with MPRester() as m:
        entries = m.get_entries_in_chemsys(chsys)
    
    pd = PhaseDiagram(entries)
    pda = PDAnalyzer(pd)
    evolution_profile = pda.get_element_profile(Element(elem), Composition(formula))
    ref = evolution_profile[0]['element_reference'].energy_per_atom
    for x in range((len(evolution_profile))):
        if abs(evolution_profile[x]['evolution']) < 0.0001:
            cat = evolution_profile[x]['chempot'] - ref
            if x + 1 == len(evolution_profile):
                ano = -20
            else:
                ano = evolution_profile[x+1]['chempot'] - ref
    
    return [cat, ano]
```

```python
def get_chempots_phases_enthalpies(chsys, formula, elem):
    chempots = [] #critical chemical potentials at every transition point
    evolutions = [] 
    phases = []
    enthalpies = []
    yvec = []
    
    with MPRester() as m:
        entries = m.get_entries_in_chemsys(chsys)
        mpentry = m.get_entries(formula) #gets all MP-entries associated with given composition
    
    comp_entry = Composition(formula)
    pd = PhaseDiagram(entries)
    pda = PDAnalyzer(pd)
    evolution_profile = pda.get_element_profile(Element(elem), comp_entry)
    ref = evolution_profile[0]['element_reference'].energy_per_atom
    
    if len(mpentry) == 0: #composition is metastable, so force it to lie on the convex hull
        compound = PDEntry(Composition(formula), 0)
        e0 = pda.get_e_above_hull(compound)
        compound = PDEntry(Composition(formula), -(e0/comp_entry.num_atoms))
    else: #otherwise return the most stable entry in the list of MP-entries. This entry can be metastable.
            #in the future, may want to include a shift of "e-above-hull" to set this compound on the x-axis
        mpentry = sorted(mpentry, key=lambda e: e.energy_per_atom)
        compound = mpentry[0]
    
    reac = [compound]
    reac.append(evolution_profile[0]['element_reference'])
    
    for stage in evolution_profile:
        chempots.append(stage['chempot'] - ref)
        evolutions.append(stage['evolution'])
        namelist = []
        for entry in stage['entries']:
            namelist.append(entry.name)
        phases.append(namelist)
    
    yshift = 0  #initialize y-shift
    window = False #assume the compound has no stability window at first
    for stage in evolution_profile:
        rxn = ComputedReaction(reac, stage['entries']) #calculate reaction enthalpy
        rxn.normalize_to(Composition(compound.name))
        enthalpies.append(rxn.calculated_reaction_energy)
        if abs(stage['evolution']) < 0.0001: #the compound has a stability window
            window = True
            yshift = stage['evolution']*stage['chempot'] + rxn.calculated_reaction_energy
    
    if not window: #if there is no stability window, check if it is metastable
        ehull = pda.get_e_above_hull(compound)
        if ehull > 0: #the compound is metastable
            yshift = -(ehull*comp_entry.num_atoms) #set y-shift to the e above hull for the formula unit
            
    for i in range(len(chempots)): #manipulate and shift data for graphs
        ynew = (evolutions[i]*(-chempots[i]) + enthalpies[i]) - yshift
        yvec.append(ynew)            
    
    return [chempots, phases, yvec, enthalpies, evolutions]
```

```python
def stability_plot(listylist, labels):
    N = len(listylist)
    length = []
    bottom = []
    for item in listylist:
        length.append(item[0]-item[1])
        bottom.append(item[1])
    length = tuple(length)
    bottom = tuple(bottom)
    ind = np.arange(N)  # the x locations for the groups
    width = 0.5       # the width of the bars

    fig, ax = plt.subplots()
    rects = ax.bar(ind + width/2., length, width, bottom, color='g')

    # add some text for labels, title and axes ticks
    ax.set_ylabel(r'$\mu_O$ (eV)')
    ax.set_title(r'Stability Window of Materials With Respect to $\mu_O$')
    ax.set_xticks(ind + width / 2)
    ax.set_xticklabels(labels)
    
    for rect in rects:
        height = rect.get_height()
        ypos = rect.get_y()+rect.get_height() + 0.05
        ax.text(rect.get_x() + rect.get_width()/4., ypos, '{:5.3f}'.format(rect.get_height()))

    return plt
```

```python
LCO = get_stability_window(['Li','Co','O'],'LiCoO2','O')
LMO = get_stability_window(['Li', 'Mn','O'],'LiMn2O4','O') #only stable at single chemical potential
LFP = get_stability_window(['Li','Fe','P','O'],'LiFePO4','O')
NMC = get_stability_window(['Li','Ni','Mn','Co','O'],'LiNiMnCoO2','O')
```

```python
plt = stability_plot([LCO, NMC, LFP],(r'LiCoO$_2$', r'LiNiMnCoO$_2$', r'LiFePO$_4$'))
plt.ylim(-4, 0.5)
plt.show()
```

# Multi-System

## Basic Interface: LCO and LTP mix

```python
def get_mix_stability(m_id1, m_id2, chemsys, elem, ratio):
    '''m_id1 and m_id2 are Materials Project material identifier strings, chemsys is a list of strings of all the 
    elements in the system, and elem is the reference element. '''
    with MPRester() as m:
        entries = m.get_entries_in_chemsys(chemsys)
        m1 = m.get_entry_by_material_id(m_id1)
        m2 = m.get_entry_by_material_id(m_id2)

    m1C = Composition(m1.name)
    m2C = Composition(m2.name)

    pd = PhaseDiagram(entries)
    pda = PDAnalyzer(pd)
    
    x = ratio
    E0 = m1.energy_per_atom * x + m2.energy_per_atom * (1 - x)
    comp = (1/m1C.num_atoms) * x * m1C + (1/m2C.num_atoms) * (1 - x) * m2C
    
    evolution_profile = pda.get_element_profile(Element(elem), comp)
    ref = evolution_profile[0]['element_reference'].energy_per_atom

    reac = [PDEntry(comp, E0)]
    reac.append(evolution_profile[0]['element_reference'])
    
    chempots = []
    evolutions = []
    phases = []
    enthalpies = []
    yvec = []
    for stage in evolution_profile:
        chempots.append(-(stage['chempot'] - ref))
        evolutions.append(stage['evolution'])
        namelist = []
        for entry in stage['entries']:
            namelist.append(entry.name)
        phases.append(namelist)
    
    yshift = 0
    for stage in evolution_profile:
        rxn = ComputedReaction(reac, stage['entries']) #calculate reaction enthalpy
        rxn.normalize_to(comp)
        enthalpies.append(rxn.calculated_reaction_energy) 
        if abs(stage['evolution']) < 0.0001: #find y-axis shift (force all compounds to be stable)
            yshift = stage['evolution']*stage['chempot'] + rxn.calculated_reaction_energy
    
    for i in range(len(chempots)): #manipulate and shift data for graphs
        ynew = (evolutions[i]*chempots[i] + enthalpies[i]) - yshift
        yvec.append(ynew)            
    
    return [chempots, phases, yvec, enthalpies]
```

```python
LTP_LCO_50 = get_mix_stability('mp-18640', 'mp-24850', ['Li','Ti','P','Co','O'], 'O', 0.5)
LTP_LCO_25 = get_mix_stability('mp-18640', 'mp-24850', ['Li','Ti','P','Co','O'], 'O', 0.25)
LTP_LCO_75 = get_mix_stability('mp-18640', 'mp-24850', ['Li','Ti','P','Co','O'], 'O', 0.75)
```

```python
LTP_LCO_0 = get_mix_stability('mp-18640', 'mp-24850', ['Li','Ti','P','Co','O'], 'O', 0)
LTP_LCO_1 = get_mix_stability('mp-18640', 'mp-24850', ['Li','Ti','P','Co','O'], 'O', 1)
```

```python
T_6 = pnds.DataFrame({'Chemical Potential':LTP_LCO_75[0],
                      'Phase Equilibria':LTP_LCO_75[1],
                     'Reaction Enthalpy':LTP_LCO_75[3],
                     'yvec':LTP_LCO_75[2]})
display(T_6)
```

```python
fig, ax = plt.subplots()

ax.set_ylabel(r'$\Delta$H of Reaction per Formula Unit (eV)')
ax.set_xlabel(r'-$\mu_O$ (eV)')

plt.plot(LTP_LCO_25[0], LTP_LCO_25[2], 'o', ls='-', label=r'LiTi$_2$(PO$_4$)$_3$ and LiCoO$_2$ 25-75 mix')
plt.plot(LTP_LCO_50[0], LTP_LCO_50[2], 'o', ls='-', label=r'LiTi$_2$(PO$_4$)$_3$ and LiCoO$_2$ 50-50 mix')
plt.plot(LTP_LCO_75[0], LTP_LCO_75[2], 'o', ls='-', label=r'LiTi$_2$(PO$_4$)$_3$ and LiCoO$_2$ 75-25 mix')
plt.plot(LTP_LCO_0[0], LTP_LCO_0[2], 'o', ls='-', label=r'LiCoO$_2$', alpha=0.2)
plt.plot(LTP_LCO_1[0], LTP_LCO_1[2], 'o', ls='-', label=r'LiTi$_2$(PO$_4$)$_3$', alpha = 0.2)

legend = ax.legend(loc='lower left')
plt.show()
```
